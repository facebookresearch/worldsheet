```python
# def plot_results(results, image_name='', exp_name=''):
#     plt.figure(figsize=(35, 5))
#     plt.subplots_adjust(wspace=0)

#     plt.subplot(1, 7, 1)
#     plt.imshow(results['orig_img_0'])
#     plt.title('input view\n')
#     plt.axis('off')

#     plt.subplot(1, 7, 2)
#     plt.imshow(results['mesh_shape_0'][..., :3])
# #     plt.imshow(results['depth_0_rec'])
#     plt.title('novel view mesh\n')
#     plt.axis('off')

#     plt.subplot(1, 7, 3)
#     plt.imshow(results['orig_img_1'])
#     plt.title('novel view (ground-truth)\n')
#     plt.axis('off')

#     plt.subplot(1, 7, 4)
#     plt.imshow(results['rgba_1_rec'][..., :3])
#     plt.title('novel view (ours w/o inpaint)\n')
#     plt.axis('off')

#     plt.subplot(1, 7, 5)
#     plt.imshow(results['rgb_1_synsin'])
#     plt.title('novel view (SynSin)\n')
#     plt.axis('off')

#     plt.subplot(1, 7, 6)
#     plt.imshow(results['rgb_1_inpaint'][..., :3])
#     plt.title('novel view (ours)\n')
#     plt.axis('off')

#     plt.subplot(1, 7, 7)
#     plt.imshow(results['rgb_1_viewappearance'])
#     plt.title('novel view (Im2Im)\n')
#     plt.axis('off')


# def plot_results(results, image_name='', exp_name=''):
#     plt.figure(figsize=(30, 5))
#     plt.subplots_adjust(wspace=0)

#     plt.subplot(1, 6, 1)
#     plt.imshow(results['orig_img_0'])
#     plt.title('input view\n')
#     plt.axis('off')

#     plt.subplot(1, 6, 2)
#     plt.imshow(results['orig_img_1'])
#     plt.title('novel view (ground-truth)\n')
#     plt.axis('off')

#     plt.subplot(1, 6, 3)
#     plt.imshow(results['rgb_1_viewappearance'])
#     plt.title('novel view (Im2Im)\n')
#     plt.axis('off')

#     plt.subplot(1, 6, 4)
#     plt.imshow(results['rgb_1_synsin'])
#     plt.title('novel view (SynSin)\n')
#     plt.axis('off')

#     plt.subplot(1, 6, 5)
#     plt.imshow(results['rgba_1_rec'][..., :3])
#     plt.title('novel view (ours w/o inpaint)\n')
#     plt.axis('off')

#     plt.subplot(1, 6, 6)
#     plt.imshow(results['rgb_1_inpaint'][..., :3])
#     plt.title('novel view (ours)\n')
#     plt.axis('off')


# def plot_results(results, image_name='', exp_name='', show_title=False):
#     plt.figure(figsize=(36, 5))
#     plt.subplots_adjust(wspace=0)
# #     plt.tight_layout()

#     plt.subplot(1, 7, 1)
#     plt.imshow(results['orig_img_0'])
#     if show_title:
#         plt.title('input view\n')
#     plt.axis('off')

#     plt.subplot(1, 7, 2)
#     plt.imshow(results['orig_img_1'])
#     if show_title:
#         plt.title('novel view (ground-truth)\n')
#     plt.axis('off')

#     plt.subplot(1, 7, 3)
#     plt.imshow(results['rgb_1_viewappearance'])
#     if show_title:
#         plt.title('novel view (Im2Im)\n')
#     plt.axis('off')

#     plt.subplot(1, 7, 4)
#     plt.imshow(results['rgb_1_synsin'])
#     if show_title:
#         plt.title('novel view (SynSin)\n')
#     plt.axis('off')

#     plt.subplot(1, 7, 5)
#     plt.imshow(results['rgb_1_inpaint'][..., :3])
#     if show_title:
#         plt.title('novel view (ours)\n')
#     plt.axis('off')

#     plt.subplot(1, 7, 6)
#     plt.imshow(results['rgba_1_rec'][..., :3])
#     if show_title:
#         plt.title('novel view (ours w/o inpaint)\n')
#     plt.axis('off')

#     plt.subplot(1, 7, 7)
#     plt.imshow(results['mesh_shape_1'][..., :3])
# #     plt.imshow(results['depth_0_rec'])
#     if show_title:
#         plt.title('novel view mesh\n')
#     plt.axis('off')


def plot_results(results, image_name='', exp_name='', show_title=False):
    plt.figure(figsize=(20, 5))
    plt.subplots_adjust(wspace=0)
#     plt.tight_layout()

    plt.subplot(1, 4, 1)
    plt.imshow(results['orig_img_0'])
    if show_title:
        plt.title('input view\n')
    plt.axis('off')

    plt.subplot(1, 4, 2)
    plt.imshow(results['orig_img_1'])
    if show_title:
        plt.title('novel view (ground-truth)\n')
    plt.axis('off')

    plt.subplot(1, 4, 3)
    plt.imshow(results['rgb_1_inpaint'][..., :3])
    if show_title:
        plt.title('novel view (ours)\n')
    plt.axis('off')

    plt.subplot(1, 4, 4)
    plt.imshow(results['rgba_1_rec'][..., :3])
    if show_title:
        plt.title('novel view (ours w/o inpaint)\n')
    plt.axis('off')
```

```python
save_dir = f'/private/home/ronghanghu/workspace/mmf_nr/save/visualization_synsin_habitat/{exp_name}/{n_iter}/{split}'
result_files = sorted(glob(os.path.join(save_dir, '*.npz')))
```

```python
for file in tqdm(result_files[0:160]):
    d = np.load(file)
    idx = int(os.path.basename(file).split('_')[1])
    results = dict(d)
#     results['rgb_1_synsin'] = skimage.io.imread(f"{synsin_dir}/{idx:08d}.png")
#     results['rgb_1_viewappearance'] = skimage.io.imread(f"{viewappearance_dir}/{idx:08d}.png")
    plot_results(results, os.path.basename(file).split('.')[0], exp_name)
    plt.savefig(f"{save_fig_dir}/{idx:04d}.png", bbox_inches="tight")
    plt.close('all')
    d.close()
```
