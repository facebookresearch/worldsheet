```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1. / (1 + np.exp(-x))

def f1(y):
    return 1. / (sigmoid(y) * 0.75 + 0.01) - 1

def f2(y):
    return 1. / (sigmoid(y - 2.8) * 10 + 0.01) - 0.1
```

```python
y = np.linspace(-5, 5, 5000)
```

```python
plt.plot(y, f1(y))
plt.xlabel('logits')
plt.ylabel('z-value')
plt.ylim([0, 10])
plt.grid()
plt.title('Our previous z-prediction function\nz = 1. / (sigmoid(y) * 0.75 + 0.01) - 1')
```

```python
plt.plot(y, f2(y))
plt.xlabel('logits')
plt.ylabel('z-value')
plt.ylim([0, 10])
plt.grid()
plt.title('Modified SynSin z-prediction function\nz = 1. / (sigmoid(y - 2.8) * 10 + 0.01) - 0.1')
```
